```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = np.zeros((100, 100), np.uint8)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.hist(image.ravel())
plt.show()
```

```python
cv2.rectangle(image, (0, 50), (100, 100), (255), -1 ) 
# -1 means fill it
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

```python
cv2.circle(image, (50,50), 25, (127), thickness=-1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
plt.hist(image.ravel())
```
