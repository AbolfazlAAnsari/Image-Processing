# Affine Transformation

Transformation that can be expressed in the form of a matrix multiplication ( linear transformation ) followed by a vector addition ( translation ). 

## Types of Affine Transformation

Rotations ( linear transformation ) <br>
Translations ( vector addition ) <br>
Scale Operations ( linear transformation ) <br>

T ( x, y ) = ( x', y' ) <br>
x' = a_00x + a_01y + b_00 <br>
y' = a_10x + a_11y + b_10 

# Loading an Image


```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
root = os.path.join(os.getcwd(), '..', 'Images')
root
```

```python
lena = cv2.imread(os.path.join(root, 'lena.png'), -1)
```

```python
lena
```

```python
rows, cols = lena.shape
rows, cols
```

```python
cv2.circle(lena, (100,100), 10, (255,0,0), -1)
cv2.circle(lena, (100,412), 10, (255,0,0), -1)
cv2.circle(lena, (412,100), 10, (255,0,0), -1)
```

```python
plt.imshow(cv2.cvtColor(lena, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
src_matrix = np.float32([[100,100],
                        [100,412],
                        [412,100]])

dst_matrix = np.float32([[100,100],
                        [100,412],
                        [412,100]])
```

```python
matrix = cv2.getAffineTransform(src_matrix, dst_matrix)
```

```python
matrix
```

```python
result = cv2.warpAffine(lena, matrix, (cols, rows))
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
src_matrix = np.float32([[100,100],
                        [100,412],
                        [412,100]])

dst_matrix = np.float32([[100,0],
                        [100,412],
                        [500,200]])
```

```python
matrix = cv2.getAffineTransform(src_matrix, dst_matrix)
```

```python
matrix
```

```python
result = cv2.warpAffine(lena, matrix, (cols, rows))
```

```python
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
